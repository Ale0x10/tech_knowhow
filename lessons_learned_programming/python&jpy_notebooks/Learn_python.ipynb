```python
# ðŸ§  EinfÃ¼hrung: Kommentare und Syntax
# In Python beginnt ein Kommentar mit #
# Es gibt keine Blockkommentare wie /* */ in C++

print("Hallo Welt!")  # Ausgabe wie in C++: cout << "Hallo Welt";

# ðŸ§® Variablen und Typen
x = 42           # int
pi = 3.1415      # float
name = "Alice"   # string
ist_aktiv = True # bool

print(type(x), type(pi), type(name), type(ist_aktiv))

# ðŸ” Kontrollstrukturen
# if / else funktioniert ohne Klammern, aber mit EinrÃ¼ckung

if x > 10:
    print("x ist grÃ¶ÃŸer als 10")
else:
    print("x ist kleiner oder gleich 10")

# ðŸ”„ Schleifen
for i in range(5):  # entspricht for(int i=0; i<5; i++)
    print(i)

# ðŸ§® Funktionen
def addiere(a, b):
    return a + b

print(addiere(3, 4))  # wie int add(int a, int b) in C++

# ðŸ§± Klassen und Konstruktoren
class Person:
    def __init__(self, name, alter):
        self.name = name
        self.alter = alter

    def vorstellen(self):
        print(f"Ich heiÃŸe {self.name} und bin {self.alter} Jahre alt.")

p = Person("Bob", 30)
p.vorstellen()

# ðŸ§µ Listen, Dictionaries, Sets
liste = [1, 2, 3]
dictionary = {"name": "Alice", "alter": 25}
menge = {1, 2, 3}

print(liste[0], dictionary["name"], 2 in menge)

# ðŸ§  Kein Zeiger-Konzept wie in C++
# Alles sind Referenzen â€“ Speicherverwaltung automatisch

a = [1, 2, 3]
b = a
b.append(4)
print(a)  # zeigt: [1, 2, 3, 4] â†’ b und a zeigen auf dasselbe Objekt

# ðŸ§ª Exception Handling
try:
    print(10 / 0)
except ZeroDivisionError:
    print("Division durch Null!")

# ðŸ“¦ Module importieren
import math
print(math.sqrt(16))  # wie #include <cmath> in C++

# ðŸ§¹ List Comprehension
quadrate = [x**2 for x in range(5)]
print(quadrate)

# ðŸ§¬ Lambda-Funktionen
f = lambda x: x * 2
print(f(5))

# ðŸ§° Dateioperationen
with open("test.txt", "w") as f:
    f.write("Hallo Datei!")

with open("test.txt", "r") as f:
    print(f.read())

# ðŸ§ª Bonus: Typisierung mit type hints
def multipliziere(a: int, b: int) -> int:
    return a * b

print(multipliziere(3, 4))
```
